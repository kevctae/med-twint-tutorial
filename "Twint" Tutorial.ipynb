## Import "Twint"

```python
import twint
import nest_asyncio

nest_asyncio.apply()
```

## Configure Search Parameters

```python
config = twint.Config()

config.Username = "elonmusk"
config.Search = "tesla"
config.Limit = 20

twint.run.Search(config)
```

## Saving Results into a "Pandas" DataFrame

```python
import pandas as pd

config = twint.Config()

config.Username = "elonmusk"
config.Search = "tesla"
config.Limit = 20

config.Pandas = True

twint.run.Search(config)
```

```python
df = twint.storage.panda.Tweets_df

display(df)
```

## Export to a ".csv"¬†File

```python
config = twint.Config()

config.Username = "elonmusk"
config.Search = "tesla"
config.Limit = 20

config.Store_csv = True
config.Output = "Output.csv"

twint.run.Search(config)
```
